```python
import pandas as pd
from google.colab import files

# Cargar el archivo de texto generado previamente
archivo_txt = '/content/personas_gym.txt'

# Leer el archivo como DataFrame
df = pd.read_csv(archivo_txt, sep='\t')

# Función para calcular el IMC
def calcular_imc(peso, altura):
    return round(peso / (altura ** 2), 1)

# Función para generar un comentario sobre el IMC
def comentario_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif 18.5 <= imc < 24.9:
        return "Peso normal"
    elif 25 <= imc < 29.9:
        return "Sobrepeso"
    else:
        return "Obesidad"

# Función para generar una sugerencia de ejercicio
def sugerir_ejercicio(row):
    # IMC
    imc = row['IMC']
    comentario = comentario_imc(imc)

    # Peso y ritmo cardiaco
    peso = row['Peso (kg)']
    ritmo_cardiaco = row['Ritmo Cardiaco (bpm)']

    # Motivo de ingreso
    motivo = row['Motivo de Ingreso']

    # Sugerencia de ejercicio basada en el comentario del IMC y otros datos
    if comentario == "Bajo peso":
        if motivo == "Perder peso":
            return "Enfocarse en ejercicios de tonificación y pesas ligeras."
        elif motivo == "Ganar masa muscular":
            return "Ejercicios de fuerza como pesas y entrenamiento de resistencia."
        else:
            return "Ejercicios suaves de bajo impacto como caminar o nadar."

    elif comentario == "Peso normal":
        if ritmo_cardiaco > 85:
            return "Ejercicios cardiovasculares para mantener la salud, como correr o ciclismo."
        else:
            return "Ejercicios combinados de fuerza y cardio, como entrenamiento en circuito."

    elif comentario == "Sobrepeso":
        if motivo == "Perder peso":
            return "Ejercicios de alto impacto como correr o clases de aeróbicos."
        else:
            return "Entrenamiento cardiovascular moderado como caminar, nadar o ciclismo."

    else:  # Obesidad
        if motivo == "Perder peso":
            return "Comenzar con ejercicios de bajo impacto, como caminar, nadar o bicicleta estática."
        else:
            return "Ejercicios de bajo impacto con enfoque en movilidad y flexibilidad."

# Calcular el IMC y agregar una columna con el resultado
df['IMC'] = df.apply(lambda row: calcular_imc(row['Peso (kg)'], row['Altura (m)']), axis=1)

# Generar un comentario sobre el IMC
df['Comentario IMC'] = df['IMC'].apply(comentario_imc)

# Generar la sugerencia de ejercicio
df['Sugerencia de Ejercicio'] = df.apply(sugerir_ejercicio, axis=1)

# Mostrar el DataFrame con los resultados
print(df[['Nombre', 'Apellido', 'IMC', 'Comentario IMC', 'Sugerencia de Ejercicio']])

# Guardar el nuevo DataFrame con las sugerencias en un nuevo archivo txt
archivo_sugerencias = '/content/personas_gym_sugerencias.txt'
df.to_csv(archivo_sugerencias, index=False, sep='\t', header=True)

# Descargar el archivo con las sugerencias
files.download(archivo_sugerencias)
```
