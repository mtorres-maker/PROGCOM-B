```python
from IPython.display import HTML

HTML('''
<!DOCTYPE html>
<html>
<head>
  <meta charset="UTF-8">
  <title>Recolector de Dulces</title>
  <style>
    canvas {
      background: black;
      display: block;
      margin: 0 auto;
      border: 3px solid orange;
    }
    body {
      text-align: center;
      font-family: "Comic Sans MS", cursive, sans-serif;
      background-color: #222;
      color: white;
    }
  </style>
</head>
<body>
  <h1>üéÉ Recolector de Dulces Halloween üéÉ</h1>
  <p>Usa las flechas ‚¨ÖÔ∏è ‚û°Ô∏è para moverte. Evita fantasmas y zombis. ¬°Recoge dulces!</p>
  <canvas id="gameCanvas" width="600" height="400"></canvas>

  <script>
    const canvas = document.getElementById("gameCanvas");
    const ctx = canvas.getContext("2d");

    let score = 0;
    let gameOver = false;

    const player = {
      x: 270,
      y: 350,
      width: 40,
      height: 40,
      color: "orange",
      dx: 0
    };

    const candies = [];
    const ghosts = [];
    const zombies = [];

    function createObject(array, color, speed, size) {
      array.push({
        x: Math.random() * (canvas.width - size),
        y: -size,
        width: size,
        height: size,
        color: color,
        speed: speed
      });
    }

    function drawRect(obj) {
      ctx.fillStyle = obj.color;
      ctx.fillRect(obj.x, obj.y, obj.width, obj.height);
    }

    function updateObjects(array, collisionCallback) {
      for (let i = array.length - 1; i >= 0; i--) {
        const obj = array[i];
        obj.y += obj.speed;
        drawRect(obj);

        if (collision(player, obj)) {
          collisionCallback(i);
        } else if (obj.y > canvas.height) {
          array.splice(i, 1);
        }
      }
    }

    function collision(a, b) {
      return a.x < b.x + b.width &&
             a.x + a.width > b.x &&
             a.y < b.y + b.height &&
             a.y + a.height > b.y;
    }

    function drawPlayer() {
      drawRect(player);
    }

    function drawScore() {
      ctx.fillStyle = "white";
      ctx.font = "20px Comic Sans MS";
      ctx.fillText("Dulces: " + score, 10, 25);
    }

    function drawGameOver() {
      ctx.fillStyle = "red";
      ctx.font = "40px Comic Sans MS";
      ctx.fillText("GAME OVER", 180, 200);
    }

    function updateGame() {
      if (gameOver) {
        drawGameOver();
        return;
      }

      ctx.clearRect(0, 0, canvas.width, canvas.height);

      // Movimiento del jugador
      player.x += player.dx;
      if (player.x < 0) player.x = 0;
      if (player.x + player.width > canvas.width) player.x = canvas.width - player.width;

      drawPlayer();
      drawScore();

      // Dulces
      if (Math.random() < 0.03) {
        createObject(candies, "magenta", 2 + Math.random()*2, 20);
      }
      updateObjects(candies, i => {
        candies.splice(i, 1);
        score += 1;
      });

      // Fantasmas
      if (Math.random() < 0.015) {
        createObject(ghosts, "white", 1.5 + Math.random(), 30);
      }
      updateObjects(ghosts, () => {
        gameOver = true;
      });

      // Zombis
      if (Math.random() < 0.01) {
        createObject(zombies, "green", 1 + Math.random()*1.5, 35);
      }
      updateObjects(zombies, () => {
        gameOver = true;
      });

      requestAnimationFrame(updateGame);
    }

    document.addEventListener("keydown", (e) => {
      if (e.key === "ArrowLeft") player.dx = -5;
      else if (e.key === "ArrowRight") player.dx = 5;
    });

    document.addEventListener("keyup", (e) => {
      if (e.key === "ArrowLeft" || e.key === "ArrowRight") player.dx = 0;
    });

    updateGame();
  </script>
</body>
</html>
''')
```
