```python
import random
import pandas as pd
from faker import Faker

fake = Faker()

class EvaluacionCamillas:
    def __init__(self, nombre, decision, camillas_disponibles):
        self.nombre = nombre
        self.decision = decision
        if self.decision == "Urgencias":
            if camillas_disponibles > 0:
                self.estado = "Urgencias"
            else:
                self.estado = "En espera"
        else:
            self.estado = self.decision

pacientes = []
camillas = 10

for _ in range(50):
    nombre = fake.name()
    decision = random.choice(["Sala COVID", "Urgencias"])
    paciente = EvaluacionCamillas(nombre, decision, camillas)
    if paciente.estado == "Urgencias":
        camillas -= 1
    pacientes.append({
        "Nombre": paciente.nombre,
        "Recomendación": paciente.decision,
        "Diagnóstico Final": paciente.estado
    })

df_1e = pd.DataFrame(pacientes)
df_1e.to_csv("evaluacion_camillas.csv", index=False)
df_1e.head()
```

```python
import random
import pandas as pd

class ActuadorA:
    def __init__(self):
        self.registros = []

    def medir_y_actuar(self):
        for i in range(20):
            humedad = round(random.uniform(0, 10), 2)
            accion = "Abrir grifo" if humedad < 3 else "Sin acción"
            self.registros.append({
                "Medición": i + 1,
                "Humedad": humedad,
                "Acción": accion
            })

# Ejecución
actuador = ActuadorA()
actuador.medir_y_actuar()
df_3a = pd.DataFrame(actuador.registros)
df_3a.to_csv("actuador_abre_grifo.csv", index=False)
df_3a.head()
```

```python
import random
import pandas as pd

class ActuadorB:
    def __init__(self):
        self.grifo_abierto = True
        self.registros = []

    def medir_y_actuar(self):
        for i in range(20):
            humedad = round(random.uniform(0, 10), 2)
            if 3 <= humedad <= 8 and self.grifo_abierto:
                accion = "Cerrar grifo"
                self.grifo_abierto = False
            else:
                accion = "Sin acción"
            self.registros.append({
                "Medición": i + 1,
                "Humedad": humedad,
                "Acción": accion
            })

# Ejecución
actuador_b = ActuadorB()
actuador_b.medir_y_actuar()
df_3b = pd.DataFrame(actuador_b.registros)
df_3b.to_csv("actuador_cierra_grifo.csv", index=False)
df_3b.head()
```

```python
import random
import pandas as pd

class ActuadorC:
    def __init__(self):
        self.grifo_abierto = True
        self.registros = []

    def medir_y_actuar(self):
        for i in range(20):
            humedad = round(random.uniform(0, 10), 2)
            if humedad > 8 and self.grifo_abierto:
                accion = "Cerrar grifo y activar sistema de absorción"
                self.grifo_abierto = False
            else:
                accion = "Sin acción"
            self.registros.append({
                "Medición": i + 1,
                "Humedad": humedad,
                "Acción": accion
            })

# Ejecución
actuador_c = ActuadorC()
actuador_c.medir_y_actuar()
df_3c = pd.DataFrame(actuador_c.registros)
df_3c.to_csv("actuador_absorcion.csv", index=False)
df_3c.head()
```

```python
import random
import pandas as pd

class DolarDia:
    def __init__(self, base=4000, num_mediciones=20):
        self.base = base
        self.num_mediciones = num_mediciones
        self.registros = []

    def simular(self):
        valor_actual = self.base
        # Generar 20 mediciones con cambio aleatorio entre -2000 y +2000
        for i in range(1, self.num_mediciones + 1):
            cambio = random.randint(-2000, 2000)
            valor_actual += cambio
            if valor_actual > 5000:
                mensaje = "Subida significativa del dólar"
            elif valor_actual < 3000:
                mensaje = "Bajada significativa del dólar"
            else:
                mensaje = "Estable"
            self.registros.append({
                "Medición": i,
                "Valor": valor_actual,
                "Mensaje": mensaje
            })

        # Comunicado de cierre del día
        # Comparamos la última medición con la anterior
        if self.num_mediciones >= 2:
            anterior = self.registros[-2]["Valor"]
        else:
            anterior = self.base
        cierre_estado = (
            "cerró al alza" if valor_actual > anterior
            else "cerró a la baja" if valor_actual < anterior
            else "no varió"
        )
        self.registros.append({
            "Medición": "Cierre",
            "Valor": valor_actual,
            "Mensaje": f"El dólar {cierre_estado} en {valor_actual}"
        })

# ——— Ejecución ———
simulador = DolarDia()
simulador.simular()

# Crear DataFrame y guardarlo
df_4 = pd.DataFrame(simulador.registros)
df_4.to_csv("dolar_dia.csv", index=False)

# Mostrar resultados
df_4
```
