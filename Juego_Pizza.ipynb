```python
import random
import time

class Cliente:
    def __init__(self):
        self.ingredientes_posibles = ["ğŸ… Tomate", "ğŸ§€ Queso", "ğŸ„ ChampiÃ±ones", "ğŸ¥“ Tocino", "ğŸŒ¶ï¸ JalapeÃ±os", "ğŸ PiÃ±a", "ğŸ— Pollo", "ğŸ«‘ PimentÃ³n"]
        self.pedido = random.sample(self.ingredientes_posibles, 3)

    def mostrar_pedido(self):
        print("ğŸ§¾ El cliente quiere una pizza con:")
        for ing in self.pedido:
            print(f"- {ing}")

class JuegoPizzeria:
    def __init__(self):
        self.vidas = 5
        self.rondas = 10
        self.ronda_actual = 0

def generar_opciones(self, pedido_correcto, ingredientes_posibles):
    opciones = [pedido_correcto]
    while len(opciones) < 3:
        opcion = random.sample(ingredientes_posibles, len(pedido_correcto))
        if opcion not in opciones:
            opciones.append(opcion)
    random.shuffle(opciones)  # para que la correcta no estÃ© siempre primera
    return opciones

    def jugar(self):
        print("ğŸ• Â¡Bienvenido a PizzerÃ­a del Caos! ğŸ•")
        print("Prepara las pizzas correctamente segÃºn los pedidos de los clientes.\n")

        while self.vidas > 0 and self.ronda_actual < self.rondas:
            self.ronda_actual += 1
            print(f"\nğŸ”„ Ronda {self.ronda_actual}")
            cliente = Cliente()
            cliente.mostrar_pedido()
            opciones = self.generar_opciones(pedido_correcto, cliente.ingredientes_posibles)


            for i, opcion in enumerate(opciones):
                print(f"{i + 1}. {' | '.join(opcion)}")

            inicio = time.time()
            try:
                eleccion = int(input("Selecciona la opciÃ³n correcta (1-3): ")) - 1
            except ValueError:
                eleccion = -1

            tiempo = time.time() - inicio

            if tiempo > 30:
                print("â° Â¡Tardaste demasiado! Perdiste una vida.")
                self.vidas -= 1
            elif eleccion >= 0 and eleccion < 3 and opciones[eleccion] == cliente.pedido:
                print("âœ… Â¡Correcto! El cliente estÃ¡ feliz.")
            else:
                print("âŒ Â¡Incorrecto! El cliente se enojÃ³. Perdiste una vida.")
                print("La respuesta correcta era:", ' | '.join(cliente.pedido))
                self.vidas -= 1

            print(f"â¤ï¸ Vidas restantes: {self.vidas}")

        if self.vidas > 0:
            print("\nğŸ‰ Â¡Felicidades! Has completado el dÃ­a en la pizzerÃ­a.")
        else:
            print("\nğŸ’€ Has perdido todas tus vidas. Intenta de nuevo maÃ±ana.")

if __name__ == "__main__":
    juego = JuegoPizzeria()
    juego.jugar()
```

```python
import random

ingredientes = ["ğŸ… Tomate", "ğŸ§€ Queso", "ğŸ„ ChampiÃ±ones", "ğŸŒ¶ï¸ Chile", "ğŸ– Carne", "ğŸ«‘ Pimiento"]

vidas = 3
puntaje = 0
objetivo = 5

def pedir_pizza():
    # Selecciona 2 ingredientes random para la pizza
    pizza = random.sample(ingredientes, 2)
    return pizza

def mostrar_ingredientes():
    print("Ingredientes disponibles:")
    for i, ing in enumerate(ingredientes, 1):
        print(f"{i}. {ing}")
    print()

print("Â¡Bienvenido al juego de la pizzerÃ­a! ğŸ•")
print(f"Tienes {vidas} vidas â¤ï¸â¤ï¸â¤ï¸")
print(f"Debes acertar {objetivo} pedidos para ganar.\n")

while vidas > 0 and puntaje < objetivo:
    pizza = pedir_pizza()
    print(f"El cliente pide una pizza con: {pizza[0]} y {pizza[1]}\n")

    mostrar_ingredientes()

    respuesta = input("Selecciona los nÃºmeros de los ingredientes separados por espacio (ejemplo: 1 2): ").split()

    try:
        seleccion = [ingredientes[int(num)-1] for num in respuesta]
    except (ValueError, IndexError):
        print("Entrada invÃ¡lida. Por favor ingresa nÃºmeros vÃ¡lidos.\n")
        continue

    if set(seleccion) == set(pizza):
        puntaje += 1
        print(f"Â¡Correcto! Llevas {puntaje} pedidos acertados.\n")
    else:
        vidas -= 1
        print(f"Â¡Error! Te quedan {vidas} vidas.\n")

if puntaje == objetivo:
    print("ğŸ‰ Â¡Ganaste! Â¡Eres un maestro pizzero! ğŸ•ğŸ•ğŸ•")
else:
    print("ğŸ’” Perdiste todas tus vidas. Â¡Intenta otra vez!")
```
